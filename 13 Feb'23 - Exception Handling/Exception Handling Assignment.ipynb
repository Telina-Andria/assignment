## Q1. Explain why we have to use the Exception class while creating a Custom Exception.


We have to use the Exception class when creating a custom exception in Python because the Exception class is a superclass that inherits all error classes in Python.

## Q2. Write a python program to print Python Exception Hierarchy

```python
import inspect 
print("Python Exception Hierarchy:")
inspect.getclasstree(inspect.getmro(Exception))

def classtree(cls, space=0):
    print("."*space,cls.__name__)
    for subcls in cls.__subclasses__():
        classtree(subcls,space+3)
classtree(Exception)
```

## Q5. Explain ImportError. What is ModuleNotFoundError?


ImportError occurs when an imported class is not accessible or is in circular dependency.

ModuleNotFoundError is a subclass of the ImportError exception that is raised when a module cannot be found. This error can occur for several reasons, such as when the module is not installed on your computer, when the module name is misspelled or incorrectly capitalized, or when the module is located in the wrong path.
