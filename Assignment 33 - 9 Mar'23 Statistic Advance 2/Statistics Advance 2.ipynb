# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.


In statistics, the Probability Mass Function (PMF) and Probability Density Function (PDF) are both types of Probability Distribution Functions that describe the behavior of a random variable. The PMF is used for discrete variables, while the PDF is used for continuous variables. In other words, the PMF gives the probability of a random discrete variable being equal to a specific value, while the PDF gives the probability of a continuous variable being within a certain range of values.

For example, to illustrate the difference between discrete and continuous variables, let's roll a fair dice with 6 faces and plot its PMF, and then plot the PDF of the height of a population.


```python
import numpy as np
import matplotlib.pyplot as plt

# Number of flips
n_flips = 1000

# Create a list to store the results of the flips
results = []

for i in range(n_flips):
    # Generate a random number from 1 to 6
    result = np.random.choice([1, 2, 3, 4, 5, 6])
    results.append(result)

# Count the number of times each result appears
counts = np.unique(results, return_counts=True)

# Get PMF
pmf = counts[1] / n_flips

# Create a bar chart of the results
plt.bar(counts[0], pmf)
plt.xlabel("Result")
plt.ylabel("Frequency")
plt.title("Probability Mass Function of Rolling a dice 1000 times")
plt.show()
```

```python
import seaborn as sns

# Mean height
mean_height = 170

# Standard deviation
std_dev = 10

# Generate 1000 heights
heights = np.random.normal(mean_height, std_dev, 1000)

# Create a histogram of the heights with PDF
sns.histplot(heights, kde=True)
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Heights")
plt.show()
```

# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?



A cumulative distribution function (CDF) describes the probabilities of a random variable having values less than or equal to x. It is a cumulative function because it sums the total likelihood up to that point. Its output always ranges between 0 and 1.

CDFs have the following definition:

CDF(x) = P(X  x)

For example, if we want to calculate the probability of person to be less than or equal to 165 cm :

```python
import scipy.stats as stats

# Mean height
mean_height = 170

# Standard deviation
std_dev = 10

# Calculate the z-score of 165 cm
z = (165 - mean_height) / std_dev

# Calculate the CDF
cdf = stats.norm.cdf(z)

# Print the CDF
print(cdf)
```

# Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution

Here are some points to explain why the Normal Distribution is the most important distribution:
1. It is the most important probability distribution in statistics because it accurately describes the distribution of values for many natural phenomena. eg: height and weight of people,iq scores,test scores, measure of errors,...
2. Some statistical hypothesis tests assume that the data follow a bell curve.
3. Linear and nonlinear regression both assume that the residuals follow a Gaussian distribution. 
4. The central limit theorem states that as the sample size increases, the sampling distribution of the mean follows a normal distribution even when the underlying distribution of the original variable is non-normal.

For example, here are real-life example of Normal Distribution: 
1. Heights of people: The heights of people are normally distributed, with a mean of about 170 cm and a standard deviation of about 10 cm. This means that most people are between 160 cm and 180 cm tall, and very few people are much taller or shorter than that.
2. IQ scores: IQ scores are also normally distributed, with a mean of 100 and a standard deviation of 15. This means that about 68% of people have IQ scores between 85 and 115, and about 16% of people have IQ scores below 85 or above 115.
3. Test scores: Test scores are often normally distributed, especially if the test is well-designed. This means that the scores are evenly distributed around the mean, with most scores close to the mean and fewer scores further away from the mean.
4. Measurement errors: Measurement errors are often normally distributed. This means that most errors are small, and there are fewer errors that are large.
5. Incomes: Incomes are not normally distributed in most countries, but they are often modeled as a normal distribution. This is because the normal distribution is a good way to represent the fact that there are a few very wealthy people, a few very poor people, and most people in between.

# Q7: Explain uniform Distribution with an example

A uniform distribution is a type of probability distribution in which all outcomes are equally likely. This means that the probability of any one outcome occurring is the same as the probability of any other outcome occurring. There are two types of uniform distribution:
1. Discrete Uniform Distribution: it's a distribution that has a finite number of values defined in a specified range.

A simple example of a uniform discrete distribution is the drawing of a card from a standard deck of cards. There are 52 cards in a deck of cards, and each card has an equal probability of being drawn. So, the probability of drawing any one card is 1/52.


2. Continuous Uniform Distribution: it's a distribution that has an infinite number of values defined in a specified range.

An example of this is the probability of randomly choosing any one positive number.

# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.



_The Central Limit Theorem states that the sampling distribution of the mean of a large dataset will follow a normal distribution, regardless of whether the original population follows a normal distribution or not.

If the original population follow a normal distribution of mean  and standard deviation , then the sampling distribution of mean will follow a normal distribution with same mean  and standard deviation of /sqrt(n) where n is the size of the dataset.

While the original population don't follow normal population, n (size of dataset) should superior or equal to 30.

_The CLT is a fundamental theorem in statistics and is used in a wide variety of applications, including:

1. Estimating population mean: The CLT can be used to estimate the population mean by taking a large sample of data and calculating the sample mean. 
2. Hypothesis testing: The CLT can be used to test hypotheses about the population mean. For example, we can use the CLT to test the hypothesis that the population mean is equal to a certain value.
3. Confidence intervals: The CLT can be used to construct confidence intervals for the population mean. 

# Q10: State the assumptions of the Central Limit Theorem.


The assumptions of the Central Limit Theorem are as follows:

1. The sample size should be sufficiently large (typically n > 30).
2. The samples should be independent and identically distributed.
3. The population should have a finite variance. If the population has an infinite variance, then the Central Limit Theorem may not hold.
