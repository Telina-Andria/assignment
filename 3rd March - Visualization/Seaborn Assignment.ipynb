## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Line Plot: A line plot is used to visualize the relationship between two continuous variables over time. It is often used to show trends or changes in data over time.

Bar Plot: A bar plot is used to compare the values of different categories. It is often used to show the distribution of data across different categories.

Histogram: A histogram is used to show the distribution of a single continuous variable. It is often used to show the frequency of data within different ranges or bins.

Scatter Plot: A scatter plot is used to visualize the relationship between two continuous variables. It is often used to show the correlation or association between two variables.

Heatmap: A heatmap is used to visualize the relationship between two categorical variables. It is often used to show how different categories are related to each other.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")

sns.relplot( x ="timepoint", y = "signal", data =fmri, kind="line" )
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
sns.catplot( x="pclass", y="age", kind ="box" ,data=titanic)
```

```python
sns.set(rc={"figure.figsize":(8, 7)})
sns.boxplot(x="pclass", y="fare" , data=titanic)
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
sns.set(rc={"figure.figsize":(12, 5)})
sns.histplot( data = diamonds, x="price", hue="cut")
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset("iris")
sns.pairplot(data = iris, hue="species")
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset("flights")
df_heatmap = flights.pivot(index= "year", columns="month", values="passengers")
sns.heatmap(data=df_heatmap)
```
