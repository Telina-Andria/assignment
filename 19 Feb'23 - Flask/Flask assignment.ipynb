## Q3. What is App routing in Flask? Why do we use app routes?

App routing is the process of binding function to URL.It allows developpers to define different pages or endpoints of their web application and specify what code should be executed when user visits each page.

We use app routes in Flask for:
1) To organise: by defining separate function for each page or endpoint, we can keep our code organized and easier to maintain.
2) To create user-friendly URLS: It allow us to define custom URLs for our web appications that are easy for user to remember and share.
3) To handle user requests: When user visits a paricular URL, Flask uses the corresponding app route to execute the appropriate code and generate a response to the user's request.

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask the url_for() function is used for URL Building. It generate a URL for the specific endpoint with given method and arguments.

```python
from flask import Flask, url_for


app = Flask(__name__)

@app.route("/")
def func1():
    s="<ul><li>Company Name: ABC Corporation</li><li>Location: India</li><li>Contact Detail: 999-999-9999</li></ul>"
    return s

@app.route("/welcome")
def func2():
    return "Welcome to ABC Corporation"

with app.test_request_context():
    print(url_for("func1"))
    print(url_for("func2"))
```
